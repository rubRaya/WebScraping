## Importing Libraries & Methods

```python
import pandas as pd
from bs4 import BeautifulSoup as bs

# connect avec le site
from urllib.request import urlopen
```

* Inputting the URL

```python
url = 'https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=illustrator'
```

* Create a Client-based Request to Get the Url


```python
client = urlopen(url)
```

* Getting the HTML Code of the Full Page

```python
html= client.read()
```

```python
html
```

* Closing the request-Connection

```python
client.close()
```

* Creating an Html Parser Using BeatifulSoup

```python
soup = bs(html, "html.parser") #parser organise le code
soup
```

* Creating a Container for the Needed Data

```python
containers= soup.find_all("div", {"class":"css-1gatmva e1v1l3u10"})
```

* Bringing it All Togather

```python
f = open("C:/Users/user/Documents/wuzzuf-illustor.csv", "w") # ou open("data/wuzzuf-illustor.tsv", "w")
header= "job_title, company_name, job_type \n"             # ou "job_title\t"
f.write(header)
```

```python
for container in containers:
    jobTitle = container.findAll("h2",{"class":"css-m604qf"})
    job_title = jobTitle[0].text.strip()
    
    companName = container.findAll("a",{"class":"css-17s97q8"})
    company_name = companName[0].text.strip()
    
    jobType = container.findAll("div",{"class":"css-1lh32fc"})
    job_type = jobType[0].text.strip()

    f.write(job_title + ", " + company_name + ", " + job_type + "\n")
f.close()
```
